```python
import h5py
import numpy as np
from temporaldata import Data, Interval

# Create nested Data with Interval objects
inner_data = Data(
    train=Interval(start=np.array([0.0, 1.0]), end=np.array([0.5, 1.5])),
    test=Interval(start=np.array([2.0, 3.0]), end=np.array([2.5, 3.5])),
    domain=Interval(start=0.0, end=4.0),
)
outer_data = Data(
    splits=inner_data,
    domain=Interval(start=0.0, end=4.0),
)

# Save to HDF5
with h5py.File("test.h5", "w") as f:
    outer_data.to_hdf5(f)

# Load with lazy=False
with h5py.File("test.h5", "r") as f:
    loaded = Data.from_hdf5(f, lazy=False)
    # loaded.materialize()

# This fails after file is closed
print(type(loaded.splits.train))  # <class 'temporaldata.temporaldata.LazyInterval'>
print(loaded.splits.train.end)    # RuntimeError: Unable to synchronously get dataspace
```
